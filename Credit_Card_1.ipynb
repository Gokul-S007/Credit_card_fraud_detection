Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset in to panda data frame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# print first five rows
credit_card_data.head()
```

```python
# Print last five rows
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
# Find the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Distribution of normal transactions and fraudelent transactions
credit_card_data['Class'].value_counts()
```

Data set is unbalanced

0 ------> Normal Transaction
1 ------> Fraudelent Transaction

```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Measure of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values of for both transactions
credit_card_data.groupby('Class').mean()
```

As the dataset is highly un balanced,we need to get the sample of the data by using sampling distribution

Fraud sample --->250
so we need the legit sample of count 250

```python
# Create new variable
legit_sample=legit.sample(n=198)
```

```python
# Concatinating the two sample to get a new dataset
new_dataset=pd.concat([legit_sample,fraud])
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# Checking the nature of the dataset
new_dataset.groupby('Class').mean()
```

Nature of the dataset is not changed

Splitting the model into features and Models 

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training data and splitting data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Models training

Generally,we use logical Regression Model

```python
model=LogisticRegression(solver='lbfgs',dual=False,max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
model
```

Model Evaluation

Accuracy score 

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Training_data_accuracy :",training_data_accuracy)
```

```python
# Accuracy for testing data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Testing_data_accuracy :",testing_data_accuracy)
```
